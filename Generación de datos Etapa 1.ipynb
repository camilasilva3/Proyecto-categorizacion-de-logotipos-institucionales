### Valencia

```python

for i in [uno,dos,tres,cuatro,cinco,seis,siete,ocho,nueve,diez]:
    i.drop(i.index[88:111], inplace=True)
    percentil_95 = i["Valence"].quantile(0.95)
    print(f"Percentiles del 95% de cada logotipo: {percentil_95}")
    percentil_5 = i["Valence"].quantile(0.05)
    print(f"Percentiles del 5% de cada logotipo: {percentil_5}")

    elim95 = i["Valence"].gt(percentil_95, axis=0)
    i["Valence"][elim95] = np.nan
    print(i["Valence"].values)

    elim5 = i["Valence"].lt(percentil_5, axis=0)
    i["Valence"][elim5] = np.nan
    print(i["Valence"].values)
```
