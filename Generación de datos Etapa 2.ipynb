# Generación de datos

#### Librerias 

```python
import pandas as pd
import numpy as np
from scipy import stats 
import matplotlib.pyplot as plt
archivo = 'C:/Users/camil/Desktop/TESIS/Datos/Prueba/Etapa 2.xlsx'
uno = pd.read_excel(archivo, sheet_name='1')
dos = pd.read_excel(archivo, sheet_name ='2')
tres = pd.read_excel(archivo, sheet_name ='3')
cuatro = pd.read_excel(archivo, sheet_name ='4')
cinco = pd.read_excel(archivo, sheet_name ='5')
seis = pd.read_excel(archivo, sheet_name ='6')
siete = pd.read_excel(archivo, sheet_name ='7')
ocho = pd.read_excel(archivo, sheet_name ='8')
nueve = pd.read_excel(archivo, sheet_name ='9')
diez = pd.read_excel(archivo, sheet_name ='10')
x=np.arange(1, 10)
size = 800
page= []
imagen=[]
experiment=[]
code=[]
```

#### Data antigua

```python
g = [0, 1, 2, 3, 4]
dfa = pd.concat([uno.iloc[:, g], dos.iloc[:, g], tres.iloc[:, g], cuatro.iloc[:, g], cinco.iloc[:, g], seis.iloc[:, g], siete.iloc[:, g], ocho.iloc[:, g], nueve.iloc[:, g]
, diez.iloc[:, g]])
dfa = pd.DataFrame(dfa)
dfa
```

#### Logotipo: "Carabineros"

```python
uno.drop(uno.index[9:111], inplace=True)

custm_val1 = stats.rv_discrete(a=1, name='custm_val1',values=(x,uno["Valencia"]))
custm_dom1 = stats.rv_discrete(a=1, name='custm_dom1',values=(x,uno["Dominancia"]))
custm_act1 = stats.rv_discrete(a=1, name='custm_act1',values=(x,uno["Activacion"]))

VALR1 = custm_val1.rvs(size=size, random_state = 42)
DOMR1 = custm_dom1.rvs(size=size, random_state = 42)
ACTR1 = custm_act1.rvs(size=size, random_state = 42)

for i in range(size):
    page.append(1)
    imagen.append('Carabineros')
    experiment.append('EX3')
    code.append('EX3-01')     
df1 = pd.DataFrame()
df1 = df1.assign(Page = page, Imagen = imagen, Experiment = experiment, Code = code, Valencia = VALR1, Dominancia= DOMR1, Activación = ACTR1)
df1
```
